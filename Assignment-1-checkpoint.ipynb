# <span style="color:Green">P9: What is the correlation between temperature and precipitation? The Pearson Correlation formula:
$$r = \frac{\sum(x_i-\overline{x})(y_i-\overline{y})}{\sqrt{\sum{(x_i-\overline{x})}^2{(y_i-\overline{y})}^2}}$$
</span>

```python
#Codes of P9 here
def pearsonCorrelation(data1, data2):
    meanData1 = sum(data1) / len(data1)
    meanData2 = sum(data2) / len(data1)
    # print(meanData1)
    # print(meanData2)
    numerator = sum((data1[i] - meanData1) * (data2[i] - meanData2) for i in range(len(data1)))

    denominator_x = sum((data1[i] - meanData1) ** 2 for i in range(len(data1))) ** 0.5

    denominator_y = sum((data2[i] - meanData2) ** 2 for i in range(len(data1))) ** 0.5


    return numerator / (denominator_x * denominator_y) if denominator_x * denominator_y != 0 else 0

calculatedPearsonCorrelation = pearsonCorrelation(meanTempratures,totalPrecip)
print(f"The correlation between temperature and precipitation is {calculatedPearsonCorrelation}")
```
