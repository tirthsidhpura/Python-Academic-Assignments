# <span style="color:Green">P1: Load the dataset.</span>

```python
#Codes of P1 here

import csv

data = []  # to store CSV data

with open('e760480e-1f95-4634-a923-98161cfb02fa.csv', 'r', newline='', encoding='utf-8') as file:
    reader = csv.DictReader(file)  # Read CSV file as a dictionary
    data = list(reader)  # Store all data in a list


# print([row['icu_crci_total'] for row in data])
```

# <span style="color:Green">P2: How many unique values exist in the 'oh_region' column? What are the unique values in the 'oh_region' column? How many records exist per 'oh_region'?</span>

```python
#Codes of P2 here

import numpy as np

# creating array of oh_region
ohRegion = np.array([row['oh_region'] for row in data])

#  to find the unique value of the ohRegion
uniqueValues = np.unique(ohRegion)
uniqueValuesLength = len(uniqueValues) 



print(f"Total Unique values in oh_region is: {uniqueValuesLength}")


print("*"*80)

print(f"Unique values are {uniqueValues}")




# to store the recorded length
recordsLength = {}

# loop through each unique values
for x in uniqueValues:
    recordsLength[x] = 0
# loop and condition matching for taking out length
    for y in ohRegion:
        if x == y:
            recordsLength[x] += 1 

print("*"*80)
# To print records
for k,v in recordsLength.items():
    print(f"Region {k} has total this records: {v}")
# print(recordsLength)
```

# <span style="color:Green">P3: What is the total number of hospitalizations? What is the average number of hospitalizations per day?</span>

```python
#Codes of P3 here


#  to create an array of hospitalizations from the data
hospitalizations = np.array([int(row['hospitalizations']) for row in data])

# to get the total sum of the hospitalizations data
totalNumOFhospitalizations = np.sum(hospitalizations)
# to find the average of the hospitalizations
avgOFhospitalizations = np.mean(hospitalizations)


print(f"Total number of hospitalizations are: {totalNumOFhospitalizations}")
print("*"*60)
print(f"Average number of hospitalizations is: {avgOFhospitalizations}")

```

# <span style="color:Green">P4: What are the top 5 days with the highest number of hospitalizations?</span>

```python
#Codes of P4 here

from datetime import datetime

# To create a date array with proper date format converting it from string to date
dates = np.array([datetime.strptime(row["date"], "%Y-%m-%dT%H:%M:%S")  for row in data])

# to get the unique dates in the array with indices
uniqueDates, indices = np.unique(dates, return_inverse=True)

# creating an numpy array of zero for each hospitization dates
dailyHopitalizations = np.zeros(len(uniqueDates), dtype=int)


# channging value of dailyHospitalizations with total sum of hospitalizations from unique dates
np.add.at(dailyHopitalizations, indices, hospitalizations)

# to get the top 5 dates 
gettingTop5 = np.argsort(dailyHopitalizations)[-5:][::-1]

# to get the top 5 days
top5Days = [(uniqueDates[i], dailyHopitalizations[i]) for i in gettingTop5]

# print(top5Days)
print("Top 5 days with highest number of hospitalizations is:-")
for date, value in top5Days:
    print(f"Occur on Date: {date.strftime('%Y-%m-%d')} and number is: {value}")
```

# <span style="color:Green">P7: Are there any seasonal (Winter, Spring, Summer, Fall) patterns in COVID-19 hospitalizations? </span>

```python
#Codes of P7 here
seasons = {12: 'Winter', 1: 'Winter', 2: 'Winter',
           3: 'Spring', 4: 'Spring', 5: 'Spring',
           6: 'Summer', 7: 'Summer', 8: 'Summer',
           9: 'Fall', 10: 'Fall', 11: 'Fall'}


# created a function to map season
def getSeason(date):
    month = date.month
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

# to store value of mapping
seasons = np.array([getSeason(d) for d in dates])
uniqueSeasons = np.array(["Winter", "Spring", "Summer", "Fall"])


# to create a sum of each hospitilization data season wise by matching seasons is equal to season
seasonalSum = np.array([
    np.sum(hospitalizations[seasons == season]) for season in uniqueSeasons
])


# to create a avg of each hospitilization data season wise by matching seasons is equal to season
seasonalAverages = np.array([
    np.mean(hospitalizations[seasons == season]) for season in uniqueSeasons
])


for season, avg, sum in zip(uniqueSeasons, seasonalAverages, seasonalSum):
    print(f"{season} has {avg} hospitalizations on average with total number of patients is {sum}")

```
