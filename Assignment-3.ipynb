This dataset contains Data for the observational assessment of food safety behaviors at farmers' markets in Ontario, Canada.

# <span style="color:Green">P1: Load the dataset.</span>

```python
#Codes of P1 here
# importing pandas library
import pandas as pd
# reading data in pandas 
xlData = pd.read_excel("Farmers Market Observation-dataset.xlsx")
```

# <span style="color:Green">P2: How many unique cities are represented in the dataset? What are the top 5 cities with the most market observations?</span>

```python
#Codes of P2 here

# code to find the unique city
uniqueCities = xlData["City"].nunique()
print("Unique cities are", uniqueCities)
print('*' * 40)
# code to get the first 5 cityCounts from the column xlData
topCities = xlData["City"].value_counts().head(5)

print("Top 5 cities with the most market observations is", topCities)
```

# <span style="color:Green">P3: How many markets operate year-round? What is the distribution of market visit months? How many markets are located indoors vs. outdoors?</span>

```python
#Codes of P3 here

# code to calculate the number of markets that operate year-round based on boolean sorting
yearRound = xlData[xlData["Yearround"] == "Yes"]["market_ID"].nunique()
print(f"Markets operate on year-round is {yearRound}")
print('*' * 40)

# code to calculate the number of visits per month based on visit_month column
visitMonth = xlData.groupby("Visit_month")["market_ID"].nunique()
print("Distribution of market visit months:\n", visitMonth)

print('*' * 40)
# code to calculate the count of markets based on location column
indoorOutdoor = xlData.groupby("Location")["market_ID"].nunique()
print("Markets located indoors vs. outdoors:\n", indoorOutdoor)

```

# <span style="color:Green">P4: How many vendors were observed in total? What is the average number of vendors per market? What is the distribution of vendor roles?</span>

```python
#Codes of P4 here

# code to get the code of unique vender id to get a total observed vendor 
totalVendors = xlData["vendor_ID"].fillna(0).nunique()
print(f"Total number of vendors observed are {totalVendors}")
print('*' * 40)

avgMarketVendors = xlData.groupby("market_ID")["vendor_ID"].nunique().mean()
print(f"Average number of vendors per market are {avgMarketVendors}" )

print('*' * 40)
# Count vendor roles distribution
vendorrolesDistribution = xlData["vendor_roles"].value_counts()
print("Distribution of vendor roles:\n", vendorrolesDistribution)
```

# <span style="color:Green">P5: How many markets have communal sinks available? How many markets allow pets?</span>

```python
#Codes of P5 here


# boolean masking is done to check communal_sinks is yes and it selects market_ID column and then nunique will return unique count
communalSink = xlData[xlData["communal_sinks"] == "Yes"]["market_ID"].nunique()
print(f"Markets who had communal sinks are {communalSink}")

# boolean masking is done to check pet_signs is yes and it selects market_ID column and then nunique will return unique count
petAllowed = xlData[xlData["pet_signs"] == "Yes"]["market_ID"].nunique()
print(f"Markets who allowed pets are {petAllowed}")
```

# <span style="color:Green">P6: What percentage of vendors handle ready-to-eat meats (TCS_rtemeat > 0)? What percentage of vendors sell raw meat?</span>

```python
#Codes of P6 here

# created a boolean mask on TCS_rtemeat to get data above 0 and then got a count of unique vendor_id then to 
# retrive a percentage it is divided by totalvendors and then multiply by 100
vendorsHandlereadytoeatMeals = (xlData[xlData["TCS_rtemeat"] > 0]["vendor_ID"].nunique() / totalVendors) * 100
print(f"Percentage of vendors who handle ready to eat meals are {vendorsHandlereadytoeatMeals}")

# created a boolean mask on TCS_rawmeat to get data above 0 and then got a count of unique vendor_id then to 
# retrive a percentage it is divided by totalvendors and then multiply by 100
percentageVendorsellrawMeat = (xlData[xlData["TCS_rawmeat"] > 0]["vendor_ID"].nunique() / totalVendors) * 100
print(f"Percentage of vendors who sell raw meat are {percentageVendorsellrawMeat}")
```

# <span style="color:Green">P7: How many vendors provide food safety information? How many vendors have visible handwashing stations?</span>

```python
#Codes of P7 here


# created a boolean mask to retreice data where FS_info is yes and selected count of unique vendor_id 
foodSafetyInfo = xlData[xlData["FS_info"] == "Yes"]["vendor_ID"].nunique()
# created a boolean mask to retreice data where Visible_handwashing_station is 1 and selected count of unique vendor_id
handwashingInfo = xlData[xlData["Visible_handwashing_station"] == 1]["vendor_ID"].nunique()
print(f"Vendors who provided food safety information are {foodSafetyInfo}")
print(f"Vendors who provided visible handwashing solutiions are {handwashingInfo}")
```

# <span style="color:Green">P8: What is the average number of transactions per vendor? </span>

```python
#Codes of P8 here

transactionPerVendor = xlData.groupby("vendor_ID")["transactions"].sum().mean()
print(f"Average number of transactions per vendor is {transactionPerVendor}")
```

# <span style="color:Green">P9: What is the distribution of vendor hygiene practices (clean attire)? </span>

```python
#Codes of P9 here

# first group the data of vendor_cleanattire and then select unique vendor_id and get the count of it
vendorHygiencePractice = xlData.groupby("vendor_cleanattire")["vendor_ID"].nunique()

# print(f"Distribution of vendor hygiene practics are {vendorHygiencePractice}")

# To get the distribution divide vendorhygenepractice by totalvendors and then multiy it by 100
distribution = (vendorHygiencePractice / totalVendors) * 100
print("Distribution of vendor hygiene practics are")
print(distribution)
```

# <span style="color:Green">P10: How many vendors use gloves when handling food? What percentage of vendors attempt handwashing? What is the percentage of vendors who touch money and food?</span>

```python
#Codes of P10 here


# boolean mask is applied on Handle_gloves where it is 1 and then get a count of unique vendor_id
vendorsUsedGloves = xlData[xlData["Handle_gloves"] == 1]["vendor_ID"].nunique()
print(f"Vendors who use gloves when handling foods are {vendorsUsedGloves}")

#  to get the percentage first got the data of handwashing_attempts which is 1 and then got a percentage of it by totalvendors
perAttemptHandWashing = (xlData[xlData["handwashing_attempts"] == 1]["vendor_ID"].nunique()/ totalVendors) * 100
print(f"percentage of vendors who attempted handwashing are {perAttemptHandWashing}")


# get a data of both beh_handle_food and beh_cash which is about 0 and then to get a percentage by dividing it by totalvendors
percVend = (xlData[(xlData['beh_handle_food'] > 0) & (xlData['beh_cash'] > 0)]['vendor_ID'].nunique() / totalVendors) * 100

print(f"Percentage of vendros who touch money and food are {percVend}")
```
