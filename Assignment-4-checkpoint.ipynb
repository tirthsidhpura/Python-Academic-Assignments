# <span style="color:Green">P8: Use LinearRegression, DecisionTreeRegressor, and RandomForestRegressor to train your model. (apply 5-fold cross validation). For each regression model calculate and print RMSE score. </span>

```python
#Codes of P8 here

import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error
import numpy as np

# Defining features and target
features = ["Min Temp (°C)", "Max Temp (°C)", "Total Precip (mm)", "Spd of Max Gust (km/h)", "Total Snow (cm)"]
target = "Mean Temp (°C)"

# Select only required columns
X = csvData[features]
y = csvData[target].dropna()



# Initialize models
lr = LinearRegression()
dt = DecisionTreeRegressor()
rf = RandomForestRegressor()

# Perform 5-fold cross-validation and calculate RMSE
lr_rmse = -np.mean(cross_val_score(lr, X[:len(y)], y, cv=5, scoring='neg_root_mean_squared_error'))
dt_rmse = -np.mean(cross_val_score(dt, X[:len(y)], y, cv=5, scoring='neg_root_mean_squared_error'))
rf_rmse = -np.mean(cross_val_score(rf, X[:len(y)], y, cv=5, scoring='neg_root_mean_squared_error'))

print(f"Linear Regression RMSE: {lr_rmse:.4f}")
print(f"Decision Tree RMSE: {dt_rmse:.4f}")
print(f"Random Forest RMSE: {rf_rmse:.4f}")



```

# <span style="color:Green">P9: Calculate and print RMSE score for the test set. </span>

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X[:len(y)], y, test_size=0.2, random_state=42)

# Train models on training set
lr.fit(X_train, y_train)
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)

# Predict on test set
lr_preds = lr.predict(X_test)
dt_preds = dt.predict(X_test)
rf_preds = rf.predict(X_test)

# Calculate RMSE for test set
lr_rmse_test = np.sqrt(mean_squared_error(y_test, lr_preds))
dt_rmse_test = np.sqrt(mean_squared_error(y_test, dt_preds))
rf_rmse_test = np.sqrt(mean_squared_error(y_test, rf_preds))

print(f"Linear Regression Test RMSE: {lr_rmse_test:.4f}")
print(f"Decision Tree Test RMSE: {dt_rmse_test:.4f}")
print(f"Random Forest Test RMSE: {rf_rmse_test:.4f}")
```
